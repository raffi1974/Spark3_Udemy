```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("pyspark-basics").getOrCreate()

df = spark.read.csv("gs://bucket-name/stackoverflowposts.csv",header=True)
```

```python
df.printSchema()
```

```python
from pyspark.sql.types import IntegerType
```

```python
df_subset = df.select(["id","tags","score","view_count","creation_date"])
```

```python
df_subset.show(5)
```

```python
df_subset.withColumn("id",df_subset["id"].cast(IntegerType())).printSchema()
```

```python
df_subset = df_subset.withColumn("id",df_subset["id"].cast(IntegerType()))
```

```python
df_subset.printSchema()
```

```python
from pyspark.sql.types import TimestampType
```

```python
df_subset = df_subset.withColumn("creation_date",df_subset["creation_date"].cast(TimestampType()))
```

```python
df_subset.printSchema()
```

```python
from pyspark.sql.functions import year,month,dayofmonth
```

```python
df_subset.withColumn("creation_year",year(df_subset.creation_date)).show(5,False)
```

```python
df_subset.withColumn("creation_month",month(df_subset.creation_date)).show(10,False)
```

```python
df_subset = df_subset \
    .withColumn("creation_year",year(df_subset.creation_date)) \
    .withColumn("creation_month",month(df_subset.creation_date)) \
    .withColumn("creation_day",dayofmonth(df_subset.creation_date))
```

```python
df_subset.printSchema()
```
