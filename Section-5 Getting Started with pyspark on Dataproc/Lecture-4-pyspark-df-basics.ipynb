```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("pyspark-basics").getOrCreate()
```

```python
spark.version
```

```python
df = spark.read.csv("gs://bucket-name/stackoverflowposts.csv",header=True)
```

```python
# df.show(5,truncate=False)
```

```python
df.printSchema()
```

```python
df.select(["id","tags","score","view_count","creation_date"]).show(10,False)
```

```python
df_subset = df.select(["id","tags","score","view_count","creation_date"])
```

```python
df_subset.show(5,False)
```

```python
df_subset.filter(df_subset.score==1).show(10,False)
```

```python
df_subset.filter(
    (df.score==1)
    & 
    (df.view_count>100)    
).show(10,False)
```

```python
df_subset.where(df.score==1).show(10)
```
