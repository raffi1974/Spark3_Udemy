```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("pyspark-basics").getOrCreate()

df = spark.read.parquet("gs://bucket-name/bikeshare_trips")
```

```python
df.printSchema()
```

```python
df.registerTempTable("bikeshare_trips_tmp")

# Use createOrReplaceTempView instead of registerTempTable
df.createOrReplaceTempView("bikeshare_trips_tmp")
```

```python
qry = """
        select 
            subscriber_type,
            start_station_name,
            end_station_name
        from 
            bikeshare_trips_tmp
        limit 100
      """

spark.sql(qry).show(10,False)
```

```python
df.write.mode("overwrite").saveAsTable("bikeshare_trips_managed")
```

```python
spark.sql("select * from bikeshare_trips_managed limit 100").show(10,False)
```

```python
df.write.mode("overwrite") \
        .option("path","gs://bucket_name/bikeshare-trips-unmanaged") \
        .saveAsTable("bikeshare_trips_unmanaged")
```

```python
spark.sql("show tables").show(10,False)
```

```python
spark.sql("drop table bikeshare_trips_managed")
```
